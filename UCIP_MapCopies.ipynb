```python
from IPython.display import display
import arcgis
from arcgis.gis import GIS
from arcgis.mapping import WebMap
import os
import json

# connect to your GIS
gis = GIS("https://www.arcgis.com","USERNAME","PASSWORD")
```

```python
## Search UCIP web map

webmap_search = gis.content.search(query = 'title:UCIP_Survey_Map', item_type="Web_Map", sort_field='numViews', sort_order='desc', max_items=9999)
webmap_search
```

```python
## Select Correct search map


map_item = webmap_search[0]
map_item
```

```python
#Show layer information
webmapJSON = map_item.get_data()
for layer in webmapJSON['operationalLayers']:
    
    display(layer)
```
